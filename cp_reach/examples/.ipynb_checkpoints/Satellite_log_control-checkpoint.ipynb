## 10s Burn Simulation

```python
sat_burn = sat_sim.SatSimBurn()
t_vect = np.arange(0, 10, 0.01)
res = sat_burn.simulate(t_vect)
```

```python
def do_burn_reference(t_vect):
    sat_burn = sat_sim.SatSimBurn()
    data = []
    p_index = sat_burn.model['p_index']
    sat_burn.model['p0_dict']['w_d_x_phase'] = 0
    sat_burn.model['p0_dict']['w_d_y_phase'] = 0
    sat_burn.model['p0_dict']['w_d_z_phase'] = 0
    sat_burn.model['p0_dict']['w_d_freq'] = 0
    sat_burn.model['p0_dict']['w_d_amp'] = 0
    sat_burn.model['p0_dict']['thrust'] = thrust
    sat_burn.model['p0_dict']['kp'] = kp
    sat_burn.model['p0_dict']['kd'] = kd
    sat_burn.model['p0_dict']['kpq'] = kpq
    sat_burn.model['p0_dict']['kdq'] = kdq
    
    res = sat_burn.simulate(t_vect, integrator='rk')
    return res

def do_burn_monte_carlo(t_vect, dist, n_runs=10):
    sat_burn = sat_sim.SatSimBurn()
    data = []
    for i in range(n_runs):
        p_index = sat_burn.model['p_index']
        sat_burn.model['p0_dict']['w_d_x_phase'] = np.random.uniform(0, 2*np.pi)
        sat_burn.model['p0_dict']['w_d_y_phase'] = np.random.uniform(0, 2*np.pi)
        sat_burn.model['p0_dict']['w_d_z_phase'] = np.random.uniform(0, 2*np.pi)
        sat_burn.model['p0_dict']['w_d_freq'] = 10**np.random.uniform(-2, 1)
        sat_burn.model['p0_dict']['w_d_amp'] = np.deg2rad(dist)
        sat_burn.model['p0_dict']['thrust'] = thrust
        sat_burn.model['p0_dict']['kp'] = kp
        sat_burn.model['p0_dict']['kd'] = kd
        sat_burn.model['p0_dict']['kpq'] = kpq
        sat_burn.model['p0_dict']['kdq'] = kdq

        res = sat_burn.simulate(t_vect, integrator='rk')
        data.append(res)
    return data
```

```python
# Perform Monte Carlo Simulation for Burn Phase
t_burn_vect = np.arange(0, 20, 0.01)
data_burn_ref = do_burn_reference(t_burn_vect)
data_burn = do_burn_monte_carlo(t_vect=t_burn_vect, dist=np.deg2rad(omega_dist), n_runs=100)
```

```python
def burn_plotting(data_burn,bounds,omega_bounds):
    fig, axes = plt.subplots(5, 1, figsize=(10, 10))
    color = ['r','g','b']
    
    # Angular Velocity
    plotting.plot_burn_angular_velocity(axes[0], data_burn)
    # for i in range(3):
    #     axes[0].axhline(y=omega_bounds[i,0], color=color[i], linestyle='--')
    #     axes[0].axhline(y=omega_bounds[i,1], color=color[i], linestyle='--')

    # Attitude
    plotting.plot_burn_attitude(axes[1], data_burn)
    for i in range(3):
        axes[1].axhline(y=np.rad2deg(bounds[i+6,0]), color=color[i], linestyle='--')
        axes[1].axhline(y=np.rad2deg(bounds[i+6,1]), color=color[i], linestyle='--')
    
    # delta_vx
    plotting.plot_burn_delta_vx_error(axes[2], data_burn)
    axes[2].axhline(y=bounds[3,0], color=color[i], linestyle='--')
    axes[2].axhline(y=bounds[3,1], color=color[i], linestyle='--')

    # delta_vy
    plotting.plot_burn_delta_vy_error(axes[3], data_burn)
    axes[3].axhline(y=bounds[4,0], color=color[i], linestyle='--')
    axes[3].axhline(y=bounds[4,1], color=color[i], linestyle='--')

    
    # delta_vz
    plotting.plot_burn_delta_vz_error(axes[4], data_burn)
    axes[4].axhline(y=bounds[5,0], color=color[i], linestyle='--')
    axes[4].axhline(y=bounds[5,1], color=color[i], linestyle='--')

    
    
    axes[0].set_title('Burn Phase')
```

```python
def plot3DInvSet(points):
    plt.figure(figsize=(14,7))
    ax1 = plt.subplot(121, projection='3d', proj_type='ortho', elev=40, azim=20)
    ax1.plot3D(points[0, :], points[1, :], points[2, :],'g')
    ax1.set_xlabel('$\\theta_x$, rad')
    ax1.set_ylabel('$\\theta_y$, rad')
    ax1.set_zlabel('$\\theta_z$, rad', labelpad=1)
    ax1.set_title('Attitude Invariant Set', fontsize=20)

    plt.axis('auto')
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
def plot2DInvSet(points):
    plt.figure(figsize=(14,7))
    ax1 = plt.subplot(121)
    ax1.plot(points[0, :], points[1, :],'g')
    ax1.set_xlabel('$\\theta_x$, rad')
    ax1.set_ylabel('$\\theta_y$, rad')
    ax1.set_title('Invariant Set in Lie Algebra', fontsize=20)

    plt.axis('auto')
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```
