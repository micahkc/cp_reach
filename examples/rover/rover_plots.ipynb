## Rover reachable-set plots
Generate the EMI disturbance trajectory envelope and roll-over velocity curve using the helpers in `cp_reach.applications.rover.reachable_set`.

```python
import sys
from pathlib import Path

repo_root = None
for path in [Path.cwd().resolve()] + list(Path.cwd().resolve().parents):
    if (path / "cp_reach").exists():
        repo_root = path
        break
if repo_root is None:
    raise RuntimeError("Could not locate repo root")
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

import matplotlib.pyplot as plt
from cp_reach.applications.rover import reachable_set
```

Configure the rover parameters used by both plots. Update the values below and rerun the cells to see new results.

```python
rover_config = {
    "eps_controller": 20,
    "emi_disturbance": 20,
    "heading": 0,
    "width": 0.3,
    "COM_height": 0.06,
    "turn_radius": 8,
}
rover_config
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
reachable_set.emi_disturbance(rover_config, ax)
fig
```
